## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
df = pd.read_csv('sample_data/day.csv')
print("Preview Data:")
display(df.head())
```

**Insight:**
- Data terdiri dari 731 baris dengan 16 kolom
- Terdapat kolom season, weathersit, holiday, dan working day yang merupakan data bentuk kategori

### Assessing Data

```python
print("\nInformasi Dataset:")
df.info()

print("\nStatistik Deskriptif Dataset:")
display(df.describe())
print("\nCek Missing Values:")
print(df.isnull().sum())
print("\nCek Duplikasi:")
print(df.duplicated().sum())

print("\nUnique Values untuk Kolom Kategori:")
print("Season:", df['season'].unique())
print("Weathersit:", df['weathersit'].unique())
print("Holiday:", df['holiday'].unique())
print("Workingday:", df['workingday'].unique())
```

**Insight:**
- Tidak ditemukan adanya missing values ataupun duplikasi dalam dataset
- Kolom kategori memiliki nilai yang sesuai dengan deskripsi seperti season (1(spring), 2(summer), 3(fall), dan 4 winter) dan weathersit.

### Cleaning Data

```python
df['season'] = df['season'].map({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
df['weathersit'] = df['weathersit'].map({
    1: 'Clear',
    2: 'Mist',
    3: 'Light Rain/Snow',
    4: 'Heavy Rain/Snow'
})
df['workingday'] = df['workingday'].map({1: 'Weekday', 0: 'Weekend'})
df['holiday'] = df['holiday'].map({1: 'Holiday', 0: 'Not Holiday'})

print("Preview Data Setelah Pembersihan:")
display(df.head())
```

**Insight:**
- Kolom kategori sudah dikonversi menjadi deskripsi sehingga lebih mudah untuk dipahami
- Data sudah siap untuk ke tahap exploratory data analysis (EDA)

## Exploratory Data Analysis (EDA)

### EDA Univariate

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['cnt'], bins=30, kde=True, color="skyblue")
plt.title("Distribusi Jumlah Penyewaan Sepeda")
plt.show()
```

**Insight:**
- Distribusi cnt memiliki puncak 4000-5000

### EDA Numerik

```python
display(df.describe())
```

```python
numerik_cols = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerik_cols):
    plt.subplot(2, 3, i+1)
    sns.histplot(df[col], kde=True, color="skyblue")
    plt.title(f'Distribusi {col}')
plt.tight_layout()
plt.show()
```

```python
numerik_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 6))
sns.heatmap(numerik_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Korelasi Antar Variabel Numerik")
plt.show()
```

**Insight:**
- Memudahkan dalam melihat ringkasan statistik (mean, median, min, max, dan standar deviation)
- Menampilkan persebaran data numerik seperti temp, windspeed, dan cnt
- Memudahkan dalam mengidentifikasi apabila ada outlier atau distribusi skewed
- Pada bagian korelasi bertujuan untuk menampilkan hubungan antar variabel.

### EDA Kategorikal

```python
kategorikal_cols = ['season', 'weathersit', 'holiday', 'workingday']

plt.figure(figsize=(10, 8))
for i, col in enumerate(kategorikal_cols):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=col, data=df, palette='viridis')
    plt.title(f'Distribusi {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i, col in enumerate(kategorikal_cols):
    plt.subplot(2, 2, i+1)
    sns.barplot(x=col, y='cnt', data=df, estimator='mean', palette='coolwarm')
    plt.title(f'Rata-rata Penyewaan Sepeda Berdasarkan {col}')
    plt.ylabel("Rata-rata Penyewaan")
plt.tight_layout()
plt.show()
```

**Insight:**
- Grafik menunjukkan jumlah observasi per kategori (season, wethersit, holiday, dan workingday)
- Memudahkan dalam melihat kategori mana yang memiliki penyewaan tertinggi (rata-rata)

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='season', y='cnt', data=df, estimator=sum, palette='viridis')
plt.title("Penggunaan Sepeda Berdasarkan Musim")
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='weathersit', y='cnt', data=df, estimator=sum, palette='coolwarm')
plt.title("Penggunaan Sepeda Berdasarkan Kondisi Cuaca")
plt.show()
```

**Insight:**
- Penggunaan sepeda lebih tinggi pada weekday (hari kerja)
- pada hari libur (weekend), penggunaan sepeda cenderung lebih rendah

### Pertanyaan 2:

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='workingday', y='cnt', data=df, estimator=sum, palette='muted')
plt.title("Penggunaan Sepeda: Weekday vs Weekend")
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(x='holiday', y='cnt', data=df, estimator=sum, palette='Blues')
plt.title("Penggunaan Sepeda: Holiday vs Non-Holiday")
plt.show()
```

**Insight:**
- Musim dan cuaca mempengaruhi penggunaan sepeda secara spesifik.
- Weekday(hari kerja) memiliki penggunaan sepeda lebih tinggi dibandingkan weekend(hari libur)

## Analisis Lanjutan (Opsional)

```python
df['usage_group'] = pd.qcut(df['cnt'], q=3, labels=['Rendah', 'Sedang', 'Tinggi'])
plt.figure(figsize=(8, 5))
sns.countplot(x='usage_group', data=df, palette='muted')
plt.title("Distribusi Kelompok Penyewaan Sepeda Berdasarkan Frekuensi")
plt.xlabel("Kategori Penyewaan")
plt.show()

monthly_usage = df.groupby(['season', 'usage_group'])['cnt'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='season', y='cnt', hue='usage_group', data=monthly_usage, palette='coolwarm')
plt.title("Rata-rata Penyewaan Sepeda Berdasarkan Musim dan Kategori Penggunaan")
plt.xlabel("Musim")
plt.ylabel("Rata-rata Penyewaan")
plt.show()
```

**Insight:**
- Penggunaan sepeda paling banyak ada di Musim Fall
- Penggunaan sepeda paling sedikit ada di Musim Winter
